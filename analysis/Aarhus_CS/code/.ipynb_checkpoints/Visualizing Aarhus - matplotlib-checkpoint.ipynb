## Visualizing Aarhus Multiplex Networks with Matplotlib

Visualization is based of of the template provided by Brennan Klein. https://github.com/jkbren/matplotlib-multilayer-network

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import networkx as nx
import matplotlib.patheffects as path_effects
from mpl_toolkits.mplot3d.art3d import Line3DCollection
%matplotlib inline
```

```python
# set colors, layer names, and seed
cols = ['orange','red', 'green', 'blue', 'purple']
layer_names = ['Lunch', 'Facebook', 'Coauthor', 'Leisure', 'Work']
np.random.seed(1985)
```

```python
#read in adj matrices as networkx objects
g1 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A1.txt'))
g2 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A2.txt'))
g3 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A3.txt'))
g4 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A4.txt'))
g5 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A5.txt'))
```

```python
# set common node position based on most dense graph 
pos = nx.kamada_kawai_layout(g5)
graphs = [g1, g2, g3, g4, g5]
```

```python
w = 8
h = 4
fig, ax = plt.subplots(1, 1, figsize=(w,h), dpi=200, subplot_kw={'projection':'3d'})

for gi, G in enumerate(graphs):
    # set node positions
    ys = list(list(zip(*list(pos.values())))[0])
    zs = list(list(zip(*list(pos.values())))[1])
    xs = [gi]*len(ys) # set a common x-position of the nodes 
    
    # node colors
    cs = [cols[gi]]*len(xs)
    
    # add within-layer edges
    lines3d = [([gi]+list(pos[j]) , [gi]+list(pos[i])) for i,j in G.edges()]
    line_collection = Line3DCollection(lines3d, zorder=gi, color=cols[gi], alpha=.25, linewidths = .75)
    ax.add_collection3d(line_collection)
    
    # now add nodes
    ax.scatter(xs, ys, zs, c=cs, s = 5, edgecolors='.25', marker='o', alpha=.75, zorder=gi)
    
    # add a plane to designate the layer    
    zdiff = max(zs)-min(zs)
    ydiff = max(ys)-min(ys)
    ymin = min(ys)-ydiff*0.1
    ymax = max(ys)+ydiff*0.1
    zmin = min(zs)-zdiff*0.1 * (w/h)
    zmax = max(zs)+zdiff*0.1 * (w/h)
    zz, yy = np.meshgrid([zmin, zmax],[ymin, ymax])
    xx = np.zeros(zz.shape)+gi
    ax.plot_surface(xx, yy, zz, color=cols[gi], alpha=0.1, zorder=gi)
    
    # add label
    layertext = ax.text(gi*0.95, 1.15, -0.75, layer_names[gi], 
                        color=cols[gi], fontsize=10, zorder=1e5)

# set them all at the same x,y,zlims
ax.set_xlim(-0.1, len(graphs) - 1 + 0.1)
ax.set_ylim(min(ys)-ydiff*0.1,max(ys)+ydiff*0.1)
ax.set_zlim(min(zs)-zdiff*0.1,max(zs)+zdiff*0.1)


# select viewing angle
angle = 115
height_angle = 25
ax.view_init(height_angle, angle)

# how much do you want to zoom into the fig
ax.dist = 9
ax.set_axis_off()
ax.set_title('Aarhus CS Department', loc = 'right', y = 0.95, x = 0.9)

#save figure - png
plt.savefig('../figures/aarhus_multiplex_network.png',dpi=425,bbox_inches='tight')
plt.show()

#save figure - pdf
#plt.savefig('../figures/aarhus_multiplex_network.pdf', dpi=425,bbox_inches='tight')
#plt.show()
```
