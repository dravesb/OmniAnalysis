## Visualizing Aarhus Multiplex Network

```python
import numpy as np 
import multinetx as mx 
import matplotlib.pyplot as plt
```

```python
#Read in adjacency matrices
g1 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A1.txt'))
g2 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A2.txt'))
g3 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A3.txt'))
g4 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A4.txt'))
g5 = nx.from_numpy_matrix(np.loadtxt('../adj_mats/A5.txt'))
```

```python
#Interlayer nodes all zero
n = len(A1)
m = 5
adj_block = mx.lil_matrix(np.zeros((m*5,m*5)))
```

```python
#Define multiplex network
mg = mx.MultilayerGraph(list_of_layers=[g1,g2,g3, g4, g5],
                        inter_adjacency_matrix=adj_block)
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121)
ax1.imshow(mx.adjacency_matrix(mg,weight='weight').todense(),
		  origin='upper',interpolation='nearest',cmap=plt.cm.jet_r)
ax1.set_title('supra adjacency matrix')

ax2 = fig.add_subplot(122)
ax2.axis('off')
ax2.set_title('edge colored network')
pos = mx.get_position(mg,mx.fruchterman_reingold_layout(g2),
                      layer_vertical_shift=0.2,
                      layer_horizontal_shift=0.0,
                      proj_angle=47)
mx.draw_networkx(mg,pos=pos,ax=ax2,node_size=50,with_labels=False,
                 edge_color=[mg[a][b]['weight'] for a,b in mg.edges()],
                 edge_cmap=plt.cm.jet_r)
plt.show()
```
